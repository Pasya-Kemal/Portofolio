## Pre-Processing

```python
y = df['snow_depth']
y
```

```python
x = df.drop('snow_depth', axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("x shape:", x.shape)
print("y shape:", y.shape)
```

```python
y = y.ravel()
```

```python
print("Missing values in x:", x.isnull().sum().sum())
print("Missing values in y:", pd.isnull(y).sum())  
```

```python
x = x.fillna(x.mean())
```

```python
y = np.nan_to_num(y, nan=np.nanmean(y))
```

```python
print("Missing values in y after filling:", np.isnan(y).sum())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Modeling

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR Train MSE = ',lr_train_mse)
print('LR Train R2 = ',lr_train_r2)
print('LR Test MSE = ',lr_test_mse)
print('LR Test R2 = ',lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbr.fit(x_train, y_train)
```

```python
y_gbr_train_pred = gbr.predict(x_train)
y_gbr_test_pred = gbr.predict(x_test)
```

```python
gbr_train_mse = mean_squared_error(y_train, y_gbr_train_pred)
gbr_train_r2 = r2_score(y_train, y_gbr_train_pred)
gbr_test_mse = mean_squared_error(y_test, y_gbr_test_pred)
gbr_test_r2 = r2_score(y_test, y_gbr_test_pred)
```

```python
print('GBR Train MSE = ', gbr_train_mse)
print('GBR Train R2 = ', gbr_train_r2)
print('GBR Test MSE = ', gbr_test_mse)
print('GBR Test R2 = ', gbr_test_r2)
```

```python
gbr_results = pd.DataFrame([['Gradient Boosting', gbr_train_mse, gbr_train_r2, gbr_test_mse, gbr_test_r2]],
                           columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])
```

```python
results = pd.concat([lr_results, gbr_results], ignore_index=True)
```

```python
print(results)
```

## XGBoost

```python
import xgboost as xgb
```

```python
xg_reg = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xg_reg.fit(x_train, y_train)
```

```python
y_xg_train_pred = xg_reg.predict(x_train)
y_xg_test_pred = xg_reg.predict(x_test)
```

```python
xg_train_mse = mean_squared_error(y_train, y_xg_train_pred)
xg_train_r2 = r2_score(y_train, y_xg_train_pred)
xg_test_mse = mean_squared_error(y_test, y_xg_test_pred)
xg_test_r2 = r2_score(y_test, y_xg_test_pred)
```

```python
print('XGBoost Train MSE = ', xg_train_mse)
print('XGBoost Train R2 = ', xg_train_r2)
print('XGBoost Test MSE = ', xg_test_mse)
print('XGBoost Test R2 = ', xg_test_r2)
```

```python
xg_results = pd.DataFrame([['XGBoost', xg_train_mse, xg_train_r2, xg_test_mse, xg_test_r2]],
                          columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])
```

```python
results = pd.concat([lr_results, gbr_results, xg_results], ignore_index=True)
```

```python
print(results)
```
